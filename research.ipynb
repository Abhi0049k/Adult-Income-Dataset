**17. Replace Salary values ['<=50K', '>50K'] with 0 and 1**

```python
data['income'].unique()
```

```python
data['income'].value_counts()
```

```python
sns.countplot(x='income',data=data)
plt.show()
```

```python
def salary_data(sal):
    if sal=='<=50K':
        return 0;
    else:
        return 1
```

```python
data['encoded_salary']=data['income'].apply(salary_data)
```

```python
data.head()
```

**18. Which workclass getting the highest salary?**

```python
data.groupby('workclass')['encoded_salary'].mean().sort_values(ascending=False)
```

**19. How has better chance to get salary >50K male or female?**

```python
data.columns
```

```python
data.groupby('gender')['encoded_salary'].mean().sort_values(ascending=False)
```
